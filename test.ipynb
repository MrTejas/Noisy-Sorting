```python
import numpy as np
import matplotlib.pyplot as plt
```

### Code for generating plots

```python
def H(p):
    p[0] = 1e-12
    ans = -p*np.log(p)-(1-p)*np.log(1-p)
    ans[0] = 0
    return ans

def g(p):
    return 0.5 + np.sqrt(p*(1-p))

def D(x,y):
    return x*np.log(x/y) + (1-x)*np.log((1-x)/(1-y))

N = 1000
p = np.linspace(0,0.5,N)

C1 = 1-H(p)
C2 = 1/((1/(1-H(p)))+(1/D(p,1-p)))


plt.plot(p, C1,color='black')
plt.fill_between(p, C1, color='skyblue', alpha=0.4, label='Theorem 3')

plt.plot(p, C2,color='black')
plt.fill_between(p, C2, color='skyblue', alpha=0.9, label='Theorem 4')




plt.xlabel('p')
plt.ylabel('Capacity')
plt.title('Converse Bounds')
plt.legend()
plt.grid()
plt.show()

```

```python

N = 1000
p = np.linspace(0,0.5,N)


A2 = 0.5*np.log(1/g(p))
A1 = 0.5*(1-H(p))


plt.plot(p, A1,color='black')
plt.fill_between(p, A1, y2=1, color='lightcoral', alpha=0.65, where=(A1 > 0), label='Theorem 1')

plt.plot(p, A2,color='black')
plt.fill_between(p, A2, y2=1, color='lightcoral', alpha=0.25, where=(A2> 0), label='Theorem 2')


plt.xlabel('p')
plt.ylabel('Capacity')
plt.title('Achievability Bounds')
plt.legend()
plt.grid()
plt.show()
```
